```python
class node:
 def __init__(self,value,next=None):
  self.value=value
  self.next=next
class ll:
  def __init__(self):
    self.head=None
  def insertbag(self,value):
    new_node=node(value)
    new_node.next=self.head
    self.head=new_node

  def display(self):
    temp=self.head
    while temp!=None:
     print(temp.value,end=" ")
     temp=temp.next
ob=ll()
ob.insertbag(7)
ob.insertbag(20)
ob.insertbag(50)
ob.insertbag(70)
ob.insertbag(12)
ob.insertbag(17)
ob.display()
```

```python
from collections import deque
class stack:
  def __init__(self):
    self.stk=deque()

  def push(self,value):
    self.stk.append(value)

  def pop(self):
    if len(self.stk)==0:
      return"stack underflow"
    print("removing:",self.stk.pop())

  def peek(self):
    print("top elemnt on the stack: ",self.stk[-1])

obj=stack()
obj.push(20)
obj.push(15)
obj.push(5)
obj.peek()
```

```python
from collections import deque

class Queue:
    def __init__(self):
        self.q = deque()

    def enqueue(self, value):
        # Insert at rear
        self.q.append(value)

    def dequeue(self):
        if len(self.q) == 0:
            return "empty queue"
        # Remove from front
        return self.q.popleft()

    def peek(self):
        if len(self.q) == 0:
            return "empty queue"
        # Front element
        return self.q[0]


# Testing
ob = Queue()
ob.enqueue(10)
ob.enqueue(50)
print(ob.dequeue())  # 10
ob.enqueue(80)
print(ob.peek())     # 50
```
